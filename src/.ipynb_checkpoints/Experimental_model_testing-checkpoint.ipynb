```python
from __future__ import print_function

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader

from PIL import Image
import matplotlib.pyplot as plt

import torchvision.transforms as transforms
import torchvision.models as models
from torchvision.utils import draw_bounding_boxes
from torchvision.transforms import ConvertImageDtype
import torchvision.models.detection as detection

import copy

from dataloader import CocoDataset

import pandas as pd
import numpy as np

from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

import json

from engine import train_one_epoch, evaluate
import utils
```

```python
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')
    
print(device)
```

```python
# draw a torch tensor
def show(imgs):
    if not isinstance(imgs, list):
        imgs = [imgs]
    fig, axs = plt.subplots(ncols=len(imgs), squeeze=False, figsize=(20, 14))
    for i, img in enumerate(imgs):
        img = img.detach()
        img = transforms.functional.to_pil_image(img)
        axs[0, i].imshow(np.asarray(img))
        axs[0, i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
```

```python
def sample_transform(sample):
  # convert channel last to channel first format
  img = torch.tensor(sample['img'].transpose(2, 0, 1))
  boxes = torch.tensor(sample['annot'][:, :4])
  labels = torch.tensor(sample['annot'][:, 4].astype(int))
  img_id = torch.tensor(sample['img_id'])

  boxes[:, 2] = boxes[:, 0] + boxes[:, 2]
  boxes[:, 3] = boxes[:, 1] + boxes[:, 3]


  annot = {'boxes': boxes, 'labels': labels, 'image_id': img_id}
  return (img, annot)
  
coco_dataset = CocoDataset('../miniCOCO', 'minitrain2017', transform=sample_transform)
```

```python
# Load miniCOCO label dictionary
label_dict = pd.read_csv('../miniCOCO/annotations/class_labels.csv', header=None, index_col=1, squeeze=True).to_dict()
label_dict = {coco_dataset.label_to_coco_label(l): v for l, v in label_dict.items() if l != 80}
print(label_dict)
```

```python
class ModelWrapper(nn.Module):
  def __init__(self, model):
    super(ModelWrapper, self).__init__()
    self.model = model

  def forward(self, input, targets=None):
    results = []
    for r in self.model(input, targets):
      results.append((r['boxes'], r['labels'], r['scores']))
    return results
```

```python
model = ModelWrapper(models.detection.fasterrcnn_mobilenet_v3_large_fpn(pretrained=True, pretrained_backbone=True)).to(device)
model.eval()
```

```python
coco_dataloader = DataLoader(coco_dataset, batch_size=1, shuffle=True, collate_fn=utils.collate_fn)
img, annot_gt = next(iter(coco_dataloader))
img, annot_gt = (img[0], annot_gt[0])

boxes, labels, scores = model([img.to(device)])[0]

annot_model = torch.zeros(1, 4, device=device)
named_labels = []

for i in range(len(labels)):
  b = boxes[i, :]
  l = labels[i].item()
  s = scores[i].item()

  if s >= 0.5:
    annot_model = torch.cat((annot_model, b.reshape(1, 4)), 0)
    named_labels.append(label_dict[l])

annot_model = annot_model[1:, :]


img = ConvertImageDtype(torch.uint8)(img.squeeze(0))
annot_img = draw_bounding_boxes(img, annot_model, named_labels)
show(annot_img)

bbox = annot_gt['boxes']
labels = [label_dict[int(l.item())] for l in annot_gt['labels']]

annot_img = draw_bounding_boxes(img, bbox, labels)
show(annot_img)

```

```python
model_experiment = models.detection.fasterrcnn_mobilenet_v3_large_fpn(pretrained=False, pretrained_backbone=True).to(device)
```

```python
# split the dataset in train and test set
indices = torch.randperm(len(coco_dataset)).tolist()
dataset = torch.utils.data.Subset(coco_dataset, indices[:-50])
dataset_test = torch.utils.data.Subset(coco_dataset, indices[-50:])
```

```python
def train_and_eval(model, dataset, dataset_test, epochs=2, start_epoch=0):
    # define training and validation data loaders
    data_loader = torch.utils.data.DataLoader(
        dataset, batch_size=2, shuffle=True, num_workers=4,
        collate_fn=utils.collate_fn)

    data_loader_test = torch.utils.data.DataLoader(
        dataset_test, batch_size=1, shuffle=False, num_workers=4,
        collate_fn=utils.collate_fn)

    # construct an optimizer
    params = [p for p in model_experiment.parameters() if p.requires_grad]
    optimizer = torch.optim.SGD(params, lr=0.005,
                                momentum=0.9, weight_decay=0.0005)
    # and a learning rate scheduler
    lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer,
                                                    step_size=3,
                                                    gamma=0.1)

    # let's train it for 10 epochs
    num_epochs = epochs
    for epoch in range(num_epochs):
        # train for one epoch, printing every 10 iterations
        train_one_epoch(model_experiment, optimizer, data_loader, device, epoch + start_epoch, print_freq=500)
        # update the learning rate
        lr_scheduler.step()
        # evaluate on the test dataset
        evaluate(model_experiment, coco_dataset.coco, data_loader_test, device=device)

train_and_eval(model_experiment, dataset, dataset_test, 10)
torch.save(model_experiment.state_dict(), '../model_experiment.backup')
```
